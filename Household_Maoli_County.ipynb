```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

from unicodedata import normalize



table = pd.read_html('https://mlhr.miaoli.gov.tw/tables2.php?y=111&m=4&unit=')

plt.rcParams['font.family'] = 'Microsoft JhengHei'

plt.rcParams['font.size'] = 12

Table = table[0].iloc[:18,[0,3,4,5]]

Table.columns = ['City', 'Male', 'Female', 'Total']

Table.loc[:, "Male"] = pd.to_numeric(Table.loc[:, "Male"])

Table.loc[:, "Female"] = pd.to_numeric(Table.loc[:, "Female"])

Table.loc[:, "Total"] = pd.to_numeric(Table.loc[:, "Total"])

plt.figure(figsize = (16,16)) # figure size

plt.subplot(2,2,1)

plt.plot(Table.City,Table.Male, label = 'Male')

plt.plot(Table.City,Table.Female, label = 'Female')

plt.plot(Table.City,Table.Total, label = 'Total')

plt.xticks(rotation = 90)

plt.legend()



Table3 = table[0].iloc[:18,[6,7]]

Table3.columns = ['Inmigration','Outmigration']

plt.subplot(2,2,2)

plt.plot(Table.City,Table3.Inmigration, label = 'In-migration')

plt.plot(Table.City,Table3.Outmigration, label = 'Out-migration')

plt.xticks(rotation = 90)

plt.legend()



Table1 = table[0].iloc[:18,2]

Table1.columns = ['number of households']

plt.subplot(2,2,3)

plt.bar(Table.City,Table1)

plt.xticks(rotation = 90)



Table2 = table[0].iloc[:18,[8,9]]

Table2.columns = ['Birth','Died']

plt.subplot(2,2,4)

plt.plot(Table.City,Table2.Birth, label = 'Birth')

plt.plot(Table.City,Table2.Died, label = 'Died')

plt.xticks(rotation = 90)





plt.legend()

plt.savefig('410821323_3-2.jpg')



```
